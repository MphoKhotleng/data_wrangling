```python
factor = dict({
    1 : 'extraversion', 
    2 : 'agreeableness',
    3 : 'conscientiousness', 
    4 : 'emotional stability',
    5 : 'intellect'})
'''
The function
takes in entry and trait as parameters
(the two values)
'''
def score(entry, trait):

    '''Find  the string of 2 values in an entry and regarding
    them as 2 columns. The score function does this by
    finding the digits in the string 
    and splitting them then regard them
    as columns._r finds the digits in the
    entry of this subcolumn'''
 
    numbers=0
    _r= re.compile('\d+')
    for col in entry:
        num= _r.findall(col)
        if str(trait)==num[0]:
            numbers+= int(num[1])
        else:
            numbers= numbers
    return numbers
factor= dict({
    1 : 'extraversion', 
    2 : 'agreeableness',
    3 : 'conscientiousness', 
    4 : 'emotional stability',
    5 : 'intellect'})
    
for i in factor.keys():
    df_person[f'{factor[i]}']= df_person.loc[:,'Section 5 of 6 [I am always prepared.]':'Section 5 of 6 [I worry about things.]'].apply(score, args=(i,) , axis=1)
    
```
